```python
import pandas as pd
import numpy as np

data = "https://raw.githubusercontent.com/coooodek/mechanical-engineering/refs/heads/main/dataset/pipeMAWP.csv"

df = pd.read_csv(data, index_col=0)

# df.head()

# 변수 지정 구간
design_pressure = float(input("설계 압력 (MPa): "))
test_temp = float(input("시수 온도 (℃) : "))

design_temp = float(input("설계 온도 (℃) : "))
material = input("재질 : ").strip()

# 온도값 보간 구간
def interpolate_stress(temp, material, table):
    if material not in table.columns:
        raise ValueError(f"{material} 데이터 내 존재하지 않는 재질입니다. 다시 입력해주세요.")
    
    available_temps = table.index
    available_stresses = table[material]

    interpolated_stress = np.interp(temp, available_temps, available_stresses) # 전,후 데이터로 선형 보간법 이용 > 선형값 아님! 보간법 사용 불가
    return interpolated_stress

try:
    stress_at_test_temp = interpolate_stress(test_temp, material, df)
    stress_at_design_temp = interpolate_stress(design_temp, material, df)
except ValueError as e:
    print(e)
    exit()

# 계산 구간
water_test_pressure = design_pressure * 1.5
result = water_test_pressure * (stress_at_test_temp / stress_at_design_temp)

print(f"기존 수압 테스트 압력 : {water_test_pressure:.2f} MPa")
print(f"온도 보정 후 수압 테스트 압력 : {result:.2f} MPa")

# print(stress_at_test_temp, stress_at_design_temp)


```
